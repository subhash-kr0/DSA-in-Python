```python
## class of nodes in a BST
## Time complexity for the insertion:
## Time complexity for the inOrder Traversal: O(n)
## Time complexity for the minimum value finding in BST:
class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data

## insertion in BST
def insertBST(root, key):
    if root is None:
        return Node(key)
    else:
        if root.data == key:
            return root
        elif root.data < key:
            root.right = insertBST(root.right, key)
        else:
            root.left = insertBST(root.left, key)
    return root

## minimum value in BST
def minValBST(node):
    temp = node
    while temp.left is not None:
        temp = temp.left
    return temp.data

## deletion of node in BST
def deleteNode(root, key):
    ## base case condition
    if root is None:
        return root
    
    ## searching of the node to be deleted
    if key < root.data:
        ## recursive call towards the left side
        root.left = deleteNode(root.left, key)
    
    elif key > root.data:
        ## recursive call towards the right side
        root.right = deleteNode(root.right, key)
        
    else:
        ## deletion part happen
        ## case 1 and case 2
        if root.left is None:
            temp = root.right
            root = None
            return temp
        
        elif root.right is None:
            temp = root.left
            root = None
            return temp
        ## case 3
        ## inorder successor
        ## minimum value in the right subtree
        temp = minValBST(root.right)
        print()
        print("Min val in the right subtree", temp)
        root.data = temp
        ## Delete the inorder successor
        root.right = deleteNode(root.right, temp)
    return root
        

## inorder traversal of BST
def inOrder(root):
    if root:
        ## recursive call for left subtree
        inOrder(root.left)
        print(str(root.data)+" ", end=' ')
        ## recursive call for right subtree
        inOrder(root.right)
        
## Driver method
root = Node(100)
root = insertBST(root, 80)
root = insertBST(root, 110)
root = insertBST(root, 50)
root = insertBST(root, 20)
root = insertBST(root, 90)

print("Inorder Traversal")
inOrder(root)

result = minValBST(root)
print()
print("Minimum value in BST", result)

## no child
deleteNode(root, 90)
print()
print('Inorder Traversal of the modified tree')
inOrder(root)

## single child
deleteNode(root, 50)
print()
print('Inorder Traversal of the modified tree')
inOrder(root)

## two child 
deleteNode(root, 100)
print()
print('Inorder Traversal of the modified tree')
inOrder(root)
```
